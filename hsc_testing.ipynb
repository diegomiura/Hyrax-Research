```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

# Load your FITS file
filename = 'images/shalo_091-503737_v3_HSC_GRIZY.fits'
hdulist = fits.open(filename)
hdulist.info()
```

```python
# Load the i-band science image
data = hdulist['SUBARU_HSC.I'].data

# Display it with a log stretch
plt.figure(figsize=(6, 6))
plt.imshow(np.log10(data + 1e-3), cmap='gray', origin='lower')
plt.colorbar(label='log(flux)')
plt.title('i-band (science image)')
plt.show()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

# Load both FITS files
idealized_file = 'images/shalo_072-0_v0_photo.fits'
survey_file = 'images/shalo_091-503737_v3_HSC_GRIZY.fits'

ideal_hdul = fits.open(idealized_file)
survey_hdul = fits.open(survey_file)

# Load the i-band images
ideal_i = ideal_hdul['SUBARU_HSC.I'].data
survey_i = survey_hdul['SUBARU_HSC.I'].data

# Plot both images with same color scaling
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

vmin = np.log10(np.percentile(ideal_i, 1) + 1e-3)
vmax = np.log10(np.percentile(ideal_i, 99) + 1e-3)

axs[0].imshow(np.log10(ideal_i + 1e-3), cmap='gray', origin='lower', vmin=vmin, vmax=vmax)
axs[0].set_title("Idealized i-band")
axs[0].axis('off')

axs[1].imshow(np.log10(survey_i + 1e-3), cmap='gray', origin='lower', vmin=vmin, vmax=vmax)
axs[1].set_title("Survey-realistic i-band")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

# Load both FITS files
idealized_file = 'images/shalo_072-0_v0_photo.fits'
survey_file = 'images/shalo_091-503737_v3_HSC_GRIZY.fits'

ideal_hdul = fits.open(idealized_file)
survey_hdul = fits.open(survey_file)

# Load the i-band images
ideal_i = ideal_hdul['SUBARU_HSC.I'].data
survey_i = survey_hdul['SUBARU_HSC.I'].data

# Plot side-by-side with independent color scaling
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Idealized plot
axs[0].imshow(np.log10(ideal_i + 1e-3), cmap='gray', origin='lower')
axs[0].set_title("Idealized i-band")
axs[0].axis('off')

# Survey-realistic plot
axs[1].imshow(np.log10(survey_i + 1e-3), cmap='gray', origin='lower')
axs[1].set_title("Survey-realistic i-band")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```
